# DBSCAN in comparision with Hierarchical and KMeans

```python
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Swapnil Mishra\Desktop\DS\DBSCAN 360 code\DBscan_clustering_codes\DBscan_clustering_codes\Dbscan.csv")
df
```

```python
plt.scatter(df.x,df.y)
plt.show()
```

### Generate clusters using Agglomerative Hierarchical CLustering 

```python
ac = AgglomerativeClustering(4,linkage = 'average')
ac_clusters = ac.fit_predict(df)
```

```python
plt.title("Clusters from Agglomerative Clustering")
plt.scatter(df.x, df.y, c= ac_clusters, s =50, cmap = 'tab20b')
plt.show()
```

### Generate clusters using KMeans Clustering

```python
km = KMeans(4)
km_clusters = km.fit_predict(df)
```

```python
plt.title("Clusters from KMeans")
plt.scatter(df.x, df.y, c = km_clusters, s =50, cmap = 'tab20b')
plt.show()
```

### Clustering using DBSCAN

```python
db = DBSCAN(eps = 0.2, min_samples=11).fit(df)
```

```python
labels = db.labels_
labels
```

```python
plt.scatter(df.x,df.y, c =labels)
plt.title("DBSCAN from scratch performance")
plt.show()
```
